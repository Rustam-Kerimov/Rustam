Система Уравнений 

```python
import numpy as np
from scipy.optimize import fsolve
def f(p):
  x,y=p
  return(x**2+y*x-10,x*y+y**2+np.sin(x)+7)
x,y= fsolve(f,(1,3))
print(x,y)

```
